```python
data_frame = train
test_data_frame = test

X = data_frame[['Age', 'Number of sexual partners', 'Num of pregnancies', 'Smokes', 'Hormonal Contraceptives', 
                'Hormonal Contraceptives (years)', 'STDs', 'STDs:condylomatosis', 'STDs:pelvic inflammatory disease', 
                'STDs:molluscum contagiosum', 'STDs:AIDS', 'STDs:HIV', 'STDs:HPV', 'Dx:Cancer', 'Dx:CIN', 'Dx:HPV']]

# label
y = data_frame[['Citology']]

testing_x = test_data_frame[['Age', 'Number of sexual partners', 'Num of pregnancies', 'Smokes', 'Hormonal Contraceptives', 
                'Hormonal Contraceptives (years)', 'STDs', 'STDs:condylomatosis', 'STDs:pelvic inflammatory disease', 
                'STDs:molluscum contagiosum', 'STDs:AIDS', 'STDs:HIV', 'STDs:HPV', 'Dx:Cancer', 'Dx:CIN', 'Dx:HPV']]
testing_y = test_data_frame[['Citology']]
```

```python
print("X size:",X.size)
print("y size:",y.size)
training_x_array = np.array(X)
training_y_array = np.array(y)
testing_x_array = np.array(testing_x)
testing_y_array = np.array(testing_y)
```

```python
model = svm.SVC()
model.fit(training_x_array, training_y_array)
model.score(testing_x_array, testing_y_array)
```
