```python
import numpy as np
import pandas as pd
from sklearn import svm
```

```python
# Spliting the data into training and testing
data = pd.read_csv('../filtered_01_cervical_cancer_dataset.csv')
```

```python
data.head()
```

```python
# we would like to shuffle the data, 
# This will make sure we have a more realistic mix of outcomes

data = data.reindex(np.random.permutation(data.index))
data.head()
```

```python
from sklearn.model_selection import train_test_split
# spliting the data, with testing set geting 30%
train, test = train_test_split(data, test_size = 0.3)
train.to_csv('training_set.csv')
test.to_csv('testing_set.csv')
```
