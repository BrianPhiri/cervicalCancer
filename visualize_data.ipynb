```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
data = pd.read_csv('filtered_01_cervical_cancer_dataset.csv')
```

```python
data.head()
```

```python
names = ['Age', 'Number of sexual partners', 'Num of pregnancies', 
         'Smokes', 'Hormonal Contraceptives', 'Hormonal Contraceptives (years)', 
         'STDs', 'STDs:condylomatosis', 'STDs:pelvic inflammatory disease', 'STDs:molluscum contagiosum', 
         'STDs:AIDS', 'STDs:HIV', 'Dx:Cancer', 'Dx:CIN', 'Dx:HPV', 'Citology']
```

```python
# Visual Correlations of the data
correlations = data.corr()

# Plot correlation matrex
# fig = plt.figure()
# ax = fig.add_subplot(111)
fig, ax = plt.subplots(figsize=(20,20))
cax = ax.matshow(correlations)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
# Scatter plot
from pandas.tools.plotting import scatter_matrix
scatter_matrix(data)
plt.show()
```
